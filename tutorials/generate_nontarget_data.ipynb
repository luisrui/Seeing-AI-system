```python
import os
import cv2 as cv
from LocateMaskFace import CropMaskFace
names = []
count = 0
```

```python
non_target_names = []
import pandas as pd
data = pd.read_csv('target_names.csv')
```

```python
target_names = list(data['0'])
```

```python
from facenet_pytorch import MTCNN
import numpy as np
mtcnn = MTCNN(margin=15, select_largest=False, post_process=False, device='cuda:0')
count = 0
for name in os.listdir('LFW_Folder/lfw_funneled'):
    print(name)
    if name not in target_names:
        pic_dir = os.path.join('LFW_Folder/lfw_funneled', name)
        for pic_name in os.listdir(pic_dir):
            img = cv.imread(os.path.join(pic_dir, pic_name))
            frame = cv.cvtColor(img, cv.COLOR_BGR2RGB)
            # Detect face
            face = mtcnn(frame)
            face = face.permute(1, 2, 0).numpy()
            face = face[...,::-1].copy()
            face = np.uint8(face)
            face, valid = CropMaskFace(face)
            print(valid)
            if valid:
                count += 1
                cv.imwrite(f'crop-mask-face/confidence/{pic_name}', face)
            break
    if count == 195:
        break
```
